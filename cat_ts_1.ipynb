```python
#http://pandas.pydata.org/pandas-docs/stable/text.html
#Creates a table by separating text with | into seperate columns
Rtypes = df['Resource Type'].str.get_dummies(sep='|')
Rtypes.head()
```

# What can we do with this information?

```python
#filtering: lets find those collections that are cartographic
df['Resource Type'].str.contains('cartographic').head()
```

```python
#nan is a problem, so lets fix that
df['Resource Type'].fillna("N/A", inplace=True)
df['Resource Type'].head()
```

```python
#All boolean now
df['Resource Type'].str.contains('cartographic').head()
```

```python
#We can use that as a filter
df['Title'][df['Resource Type'].str.contains('cartographic')]
```

```python
#What about getting modern collections?
(df['Date']>2000)
```

```python
#lets find out the data types of our data
df.dtypes
```

```python
#Lets find the modern records
df['Title'][df['Date Start']>1950]
```

```python
#lets find modern cartographic records
df['Title'][(df['Date Start']>1900) & df['Resource Type'].str.contains('cartographic')]
```
