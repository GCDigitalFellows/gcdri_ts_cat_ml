```python
import pandas as pd
df = pd.read_csv("pd_collections.csv")
```

```python
df.head(3)
```

# How do we start summarizing data?

```python
#lets order by
df.groupby("Language")
```

```python
#lets see how many of each
df.groupby('Language').count()
```

```python
#Further refine with sort 
df.groupby('Language').count().sort_values(by="Title", ascending=False)
```

```python
#Can also find largest for each catagory
df.groupby("Language").max()
```

```python
#What about numerical groups
df.groupby("Date Start").sum() 
```

```python
#Way too many, so we can refine
# simple ordering is by century - just divide by 100 and then multiply by 100 

df['Century'] = df['Date Start']//100 * 100
```

```python
# Can also use apply, which passes each row of a column into a function
# Can also be used for data cleaning functions

def get_century(date):
    # // is integer division 
    return date//100 * 100

df['Century'] = df['Date Start'].apply(get_century) 
```

```python
df['Century'].head()
```

```python
df.groupby('Century').max()
```

```python
#We can do multiple/nested grouping
df.groupby(["Language", "Century"]).max()
```

```python
#can also order by catagory
df.groupby(["Language", "Century"]).max().unstack()
```

```python
#Can also do it on just one column
df.groupby(["Language", "Century"])["Number of Items"].max().unstack()
```

```python
#or two
df.groupby(["Language", "Century"])[["Number of Items", "Genre" ]].count().unstack()
```

# What if you want to apply more than one function?

```python
# total number of items but the last start date
"""index becomes the row agg function is how 
data in each subset->(Station, Date) is analyzed:
* first argument is the column being analyzed
* second is aggregating function:
    * total items
    * max start date
"""
#now fancy grouping
df_pivot = df.pivot_table(index='Century', 
            columns='Language', 
            aggfunc={'Number of Items':'sum', 'Date Start':'max'})
df_pivot
```

```python
#flip table to make it more readable
df_pivot.T
```

```python
#can select columns from  pivot tables
df_pivot['Number of Items'].T 
```

```python
#lets get the languages that show up the most
df_pivot['Number of Items'].count().sort_values(ascending=False)
```

```python
most_popular = df_pivot['Number of Items'][["English", "French", "Latin", "Spanish"]]
most_popular.T
```

```python
#lets plot our most popular lanaguges in the collections over time
%matplotlib inline
most_popular.plot()
```

```python
#can also do bar
most_popular.plot(kind='bar')
```
