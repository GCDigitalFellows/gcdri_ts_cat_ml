```python
import pandas as pd
df = pd.read_csv("titanic.csv")
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
fig, ax = plt.subplots()

for stat, c, label in [(0,'red', 'died' ), (1,'blue', 'survived')]:
    dsub = df[df['Survived']==stat]
    ax.scatter(dsub['Age'], dsub['Fare'], s=10*(dsub['SibSp']+dsub['Parch']), c=c, label=label)
    
ax.set_title("Passengers (dot size=family size)")
ax.set_xlabel("age")
ax.set_ylabel("fare")
ax.set_xlim(0,80)
ax.set_ylim(0,200)
ax.legend(scatterpoints=1, ncol=2)
```
